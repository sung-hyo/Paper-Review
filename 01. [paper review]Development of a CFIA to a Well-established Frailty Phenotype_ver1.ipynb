## CHS의 Frailty Phenotype의 조작적 정의

Fried LP, Tangen CM, Walston J, et al. Frailty in older adults: Evidence for a phenotype. J Gerontol A Biol Sci Med Sci. 2001;56:M146–M156. [PubMed] [Google Scholar]

1. 체중감소: 전년도에 비해 10파운드(약 5kg)이상 또는 5%이상의 체중감소
    - Shrinking: weight loss, unintentional, of ≥10 pounds in prior year or, at follow-up, of ≥5% of body weight in prior year (by direct measurement of weight).

2. 성별과 BMI를 보정한 기준에서 최저 20%의 악력
    - Weakness: grip strength in the lowest 20% at baseline, adjusted for gender and body mass index.

3. 지구력과 에너지 부족: 자가 보고 피로감으로 알 수 있음. 
    - Poor endurance and energy: as indicated by self-report of exhaustion. Self-reported exhaustion, identified by two questions from the CES–D scale, is associated with stage of exercise reached in graded exercise testing, as an indicator of V̇O2 max, and is predictive of cardiovascular disease.
    * Center for Epidemiological Studies-Depression(CES–D)

4. 성별과 키를 보정한 15 feet 보행 시간으로 인구 중 가장 느린 20%를 기준으로 함
    - Slowness: The slowest 20% of the population was defined at baseline, based on time to walk 15 feet, adjusting for gender and standing height.

5. 낮은 신체활동 수준: 주 당 소비되는 kilocalories를 기준으로 하여 계산, 성별로 가장 낮은 5분위수로 정의
    - Low physical activity level: A weighted score of kilocalories expended per week was calculated at baseline, based on each participant's report. The lowest quintile of physical activity was identified for each gender.

## Abstract

### background
* Fried 등은 심혈관질환 건강연구(CHS)에서 노쇠를 표현하는 측정을 기술
* Frailty Phenotype은 다음 5가지 중 3가지가 존재할때 나타난다: 낮은 악력, 낮은 에너지(자가 보고 피로감), 느린보행속도, 낮은 신체활동, 의도하지 않은 체중감소
    - This phenotype is manifest when >= 3 of the following are present: low grip strength, low energy, slowed waking speed, low physical activity, or unintentional weight loss.
* 이 연구에서는 Frailty를 평가하는 신체적 측정(검사)없이 행정청구자료(claim data)를 이용하여 추정하고자 함

### Study design
* 참가자의 건강보험청구와 연계된 CHS코호트 자료를 이용
* 보험 청구자료를 벌점 로지스틱 회귀모형으로 분석하여 표현형 노쇠를 진단하기 위한 선형결합 형태의 간결한 지수를 개발
* 노화와 관련된 결과(입원, 장애, 사망)를 이용하여 개발된 Frailty 지수의 예측 타당성을 평가
* 기존의 정립된 Frailty Phenotype와 비교

### DATA
* 건강보험과 연계된 CHS(심혈관질환 연구)자료
    - CHS는 1989년부터 미국의 4개 지역에서 모집된 5,201명의 노인 대상 인구기반 종단 코호트(longitudinal cohort)
    - 이후 687명의 아프리카계 미국인이 추가
* 대상자는 의료금융청(Health Care Financing Administraion) 표본 추출틀에서 표본 추출한 각 가구의 모든 가구원을 포함하며, 조사 당시 65세 이상(해당 지역에 3년 이상 거주할 것으로 예상)
* 휠체어를 타거나, 호스피스 치료, 방사선 치료, 또는 암 치료용 화학요법을 받은 대상은 제외
* 코호트 대상자들은 1999년까지 매년 검사를 받았으며, 검사 후 6개월 마다 전화로 확인
* 1991년부터 2009년까지 등록된 대상자의 보험 청구 데이터를 포함
* CFI(Claims-based Frailty Indicator)를 구축하기 위해 대상자들은 청구자료를 사용한 6개월 기간 동안 의료보험 파트 A와 B를 지속 유지 

### Candidate Variable Selection(후보 변수 선택)
* 모델에 포함시킬 관련 후보변수를 식별하기 위해 pubed와 web of sicience에서 논문과 참고자료를 검색
    - 검색단어 claims based frailty measurement, frailty, railty variables, frailty index, phenotype frailty indicators,definition of frailty 
* 청구자료 또는 전자의료기록에서 Frail 또는 장애로 분류하기 위해 식별된 논문에서 사용된 임상조건을 추출
* 청구자료만으로 각 조건을 활용하는 것의 타당성을 평가하기 위해 임상조건을 전문가에게 검토 받음
* 의료 연구 및 품질기관의 임상분류 소프트웨어(CCS)의 268개 단일 레벨 진단을 사용. CCS 진단 레벨이 없는 임상 조건의 경우는 의료 코딩 소프트웨어(LLC)를 사용하여 관련 ICD9 코드로 식별
* Frailty와 상관관계가 있다고 간주되는 독립변수는 5년 또는 9년 방문 전 최대 6개월까지의 청구자료로 부터 도출
    - 각 개인은 코호트에 자료를 최대 2회 까지 제공
* 입원 및 외래환자 청구자료로 제한했고, Nursing home care 또는 숙련된 nursing care는 제외

### Modeling Frailty
* 청구자료만으로 개인의 Frailty를 예측할 수 있는 변수의 최적의 선형결합을 찾는 것
* 목표는
    - (a) Frailty에 대해 민감도와 특이도 산출
    - (b) 여러 데이터에 일반화 할수 있는 청구 기반 Frailty 지수를 개발
* 기계학습 분야에서 널리 사용되는 3가지 기술을 고려
    - logistic regression with lasso penalty, gradient boosting machine, and random forests
* 이 연구에서는 LASSO 패널티 로지싁 회귀모형을 선택하였으며, 10-fold 교차 검증으로 LASSO penalty를 최적화
* ROC의 AUC를 최대화 하는 cut-off를 선택함으로서 민간도와 특이도가 우수한 Frailty를 식별할 수 있는 예측 모델을 만드는 것
* 로지스틱 분석에서 교차 검정(cross-varidation) 기반 최적화 알고리즘은 상호 작용 (term)이 없는 알고리즘으로 모형에서는 상관관계가 있는 연령은 최종모형에서 제외 됨
* 랜덤포리스트, 그레이디언트 부스팅, 전방과 후방 변수 선택을 이용한 로지스틱 회귀분석과 동방상병질환을 사용하는 접근법등의 모델링 기법을 테스트 함
* 트리학습 알고리즘에서는 서로 다른 1차(상호작용 없음), 2차, 3차 상호작용 모형을 시도

### Assessment of Validity
* 적응형 라소 접근방식은 validation set에서 ROC 아래의 면적이 최대화 하도록 교차 검증을 수행
* 높은 양성 예측값(positive predictive)을 산출하는 컷오프(cot off)를 선택하여 CFI와 함께 Frail와 nonfrail로 분류된 개별 사건들의 비율을 조사하여 CFI의 예측 타당성을 평가
* 특이도를 최대화 하는 cut-off와 민감도와 특이도을 동시에 최대화하는 cut-off등 몇 가지 cut-off 선택하여 연구 함
* 코호트 참여 5년차 방문에서 예측된 Frailty로 개인을 Frail 및 Nonfrail로 분류하여, 이후 5년 동안 에 대한 입원, 골절, 요양원 입원, 장애(일상 생활 활동에서의 영향) 및 사망 빈도를 수량화 함
* 사망과 입원 치료는 의료 보험 청구 자료에서, 골절, 요양원 입원, 일상 생활에서의 활동 장애는 CHS 인터뷰 자료에서 사용
* 이 연구에서는 연령과 성별에 대한 보정 없이 사건의 위험비(Hazards ratios), 발생률(발병률, incidence rate ratios)을 계산
* 또한 연관성 척도(사망, 발생)의 위험 예측을 위해 Charlson Comorbidity Index만을 사용하는 척도와 비교함

### Result
* 6개월 동안 자료를 이용한 적응형 라소 회귀분석은 AUC 0.75인 노쇠 식별 모델을 산출(그림1)
    - 라소 접근법은 나이와 성별을 포함한 21개 변수를 선택(표 2)
* 21개 변수와 0.12 컷오프를 사용하여 노쇠를 분류하면 민감도와 특이도는 각각 66%, 73%였으며, 컷오프 0.20일 경우 민감도와 특이도는 각각 35%, 91%로 나타남
* 다른 모델링 기법(그레이디언트 부스팅, 랜덤포리스트)은 유사한 결과를 산출하였지만 해석하기 어려웠음(이 연구에서)
* 랜덤포리스트 AUC 0.73, 그레이디언트 부스팅 AUC 0.76, 40개 동방상병을 단순카운트하여 보정한 모델 AUC 0.72, 전후방 변수선택을 이용한 로지스틱 AUC 0.73

* 컷오프 0.20의 결과
    - 5년내 사망률(OR 3.5, CI 3.2-4.6), 사망기간(HR 3.2, CI 2.7-3.7), 5년 내 입원(OR 2.2, CI 1.7-2.9), 보정되지 않은 모델에서 요양기관 입원(OR 3.8, CI 3.0-49) (표3)

### DISCUSSION
* 개인 Frailty를 구분하는 cut off는 사용하고자 하는 의도에 따라 선택 할 수 있다. 
    - 예) Frailty를 구체적으로 구분하고 싶다면 높은 컷오프(0.25), Frailty를 보다 포괄적으로 구분하고자 한다면 0.12와 같은 컷오프가 더 적절하다. 
* 랜덤포리스트와 그레이디언트 부스팅은 간결한 모형을 얻을 수 없고, 해석이 복잡하고, 다른 이해관계자들에게 쉽게 전달할 수 없다. 
* 이 연구에서는 CHS에서 개발 된 Frailty Phenotype를 기반으로 청구자료를 이용하여 Frailty를 모델링하였다. 그러나 다른 자료에서 재현성과 동일한 변수가 선택 될 것인가?

### Zero-inflated model(0 팽창 모형)
* 0의 값이 과도하게 많은 경우(불균형 데이터)에 활용
* 0값의 관측을 가능하게(허용하는) 분포

#### Zero-inflated Possion
정해진 시간 안에 어떤 사건이 일어날 횟수에 대한 기댓값을 ${\displaystyle \lambda }$라고 했을 때, 그 사건이 ${\displaystyle n}$ 일어날 확률은 다음과 같다.
따라서 포아송 분포는 다음과 같다. 
$${\displaystyle f(n;\lambda )={\frac {\lambda ^{n}e^{-\lambda }}{n!}},\,\!} {\displaystyle f(n;\lambda )={\frac {\lambda ^{n}e^{-\lambda }}{n!}},\,\!}$$

Zero-inflated Possion은 다음과 같다.

$$
\Pr(y_{j} = 0) = \pi + (1- \pi) e^{{- \lambda}} \\
\Pr(y_{j} = h_{i}) = (1- \pi) \frac {\lambda^{h_{i}} e^{{- \lambda}}}{h_{i} !}, \qquad h_{i} \geq 1
$$

$\lambda$는 포아송 분포의 평균이고, $\pi$는 0일 확률이고, 따라서 평균은 $(1-\pi)\lambda$ 이고 분산은 $\lambda (1-\pi )(1+\pi \lambda )$.
